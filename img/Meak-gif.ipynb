```python
from PIL import Image

# لیست عکس‌ها (باید اندازه‌شون یکی باشه)
frames = [Image.open(img) for img in ["1.png", "2.png", "3.png","4.png"]]

# ذخیره به عنوان گیف
frames[0].save(
    "output1.gif",
    save_all=True,
    append_images=frames[1:],
    duration=500,  # مدت هر فریم (میلی‌ثانیه)
    loop=0         # تعداد تکرار (0 = بی‌نهایت)
)
```
